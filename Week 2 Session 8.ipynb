```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
```

```python
def sin_x(x):
    return np.sin(x)

def cos_x(x):
    return np.cos(x)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(x, sin_x(x))
ax1.set_title('sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x')

ax2.plot(x, cos_x(x))
ax2.set_title('cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x')
plt.tight_layout()
plt.savefig('sin_cos_plot.pdf')
plt.show()
```
