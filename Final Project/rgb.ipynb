```python
import astropy
from astropy.io import fits
import matplotlib.pyplot as plt
from matplotlib import rcParams
import imageio
from scipy.ndimage import zoom
%matplotlib inline

rcParams['figure.figsize'] = [10., 8.]
```

```python
r_image = fits.getdata('hlsp_hudf12_hst_wfc3ir_udfmain_f160w_v1.0_drz.fits')
g_image = fits.getdata('hlsp_hudf12_hst_wfc3ir_udfmain_f125w_v1.0_drz.fits')
b_image = fits.getdata('hlsp_hudf12_hst_wfc3ir_udfmain_f105w_v1.0_drz.fits')
```

```python
# Stack images along the third dimension
rgb_image = np.stack((r_image, g_image, b_image), axis=-1)

# Display the image
plt.imshow(rgb_image)
plt.axis('off')  # Turn off axis numbers
plt.savefig('rgb_image.png')
plt.show()
```
