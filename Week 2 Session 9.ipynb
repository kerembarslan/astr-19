```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.uniform(0, 1, 1000)
```

```python
plt.hist(random_numbers, bins=100, color='blue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Uniform Distribution [0,1] Histogram')
plt.show 
```

```python
plt.savefig('uniform_distribution_[0,1]_histogram.pdf')
```
