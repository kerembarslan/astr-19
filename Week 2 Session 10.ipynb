```python
import numpy as np
import matplotlib.pyplot as plt

random_numbers = np.random.triangular(left=1, mode=6, right=7, size=1000)

plt.hist(random_numbers, bins=100, edgecolor='black')
plt.xlabel('Days of the week')
plt.ylabel('Hours a day gaming')
plt.show() 
```
